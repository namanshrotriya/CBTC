```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Load the dataset into a DataFrame
iris_data = pd.read_excel("C:/Users/acs/OneDrive/Desktop/Projects/Python/Iris Flowers/Iris Flower.xlsx")

# Displaying the first few rows of the dataset
print("First few rows of the dataset:")
print(iris_data.head())
```

```python
# Visualizing the distribution of Sepal Length
sns.histplot(data=iris_data, x='SepalLengthCm', hue='Species', kde=True)
plt.title('Distribution of Sepal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualizing the distribution of Petal Width
sns.histplot(data=iris_data, x='PetalWidthCm', hue='Species', kde=True)
plt.title('Distribution of Petal Width by Species')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Splitting the data into training and testing sets
X = iris_data.drop('Species', axis=1)
y = iris_data['Species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Training a K-Nearest Neighbors classifier
knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train, y_train)
```

```python
# Making predictions on the test set
predictions = knn_classifier.predict(X_test)

# Evaluating the model
print("Confusion Matrix:")
print(confusion_matrix(y_test, predictions))
print("\nClassification Report:")
print(classification_report(y_test, predictions))
```
